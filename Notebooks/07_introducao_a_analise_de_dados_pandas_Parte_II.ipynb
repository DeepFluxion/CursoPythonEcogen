## Mas o que isso quer dizer:

O que é o conjunto de dados da flor Iris? É um dos conjuntos de dados mais conhecidos no mundo do aprendizado de máquina e da ciência de dados, e por um bom motivo. É composto por 150 registros de flores de íris, incluindo informações sobre o comprimento e largura das sépalas e pétalas, bem como o tipo de flor de íris. Nesta postagem do blog, exploraremos o conjunto de dados Iris e aprenderemos sobre as diferentes técnicas e métodos que podemos usar para analisá-lo e compreendê-lo. Seja você um iniciante ou um cientista de dados experiente, esta postagem fornecerá insights e dicas valiosas para trabalhar com esse conjunto de dados clássico.

## Vamos gerar estatísticas para compreender os dados usando a função `.describe()`

Assim teremos uma tabela onde para cada coluna:
- count: é a contagem dos dados
- mean : é a media
- std : é o desvio padrão
- min : é o valor mínimo
- 25 %: é o valor limite dos primeiros 25% dos dados (1o. quartil)
- 50 %: é o valor limite dos primeiros 50% dos dados (2o. quartil)
- 75 %: é o valor limite dos primeiros 75% dos dados (3o. quartil)
- max : é o valor máximo

```python
iris_data.describe()
```

```python
# Checando por missing values (valores ausentes)
iris_data.isnull().sum()
```

```python
# Checando por valores duplicados
duplicates = iris_data[iris_data.duplicated()]
print("Numero de duplicatas:", len(duplicates))
```

```python
# Drop duplicatas
iris_data = iris_data.drop_duplicates().copy()
```

```python
# Checando novamente por valores duplicados
duplicates = iris_data[iris_data.duplicated()]
print("Numero de duplicatas:", len(duplicates))
```

## Você deve ter notado que especie é uma categoria é não existe uma média, devio padrão, apenas podemos contar os valores com `.value_counts()`

```python
iris_data['Especie'].value_counts()
```

# Uma funcionalidade recente do pandas é o plot, que facilita muito a visualização dos dados.

```python
iris_data.columns
```

```python
# Visualizando com plot
iris_data['SepalaComp'].plot(figsize=(20,4));
```

```python
# Visualizando vários dados com plot
cols  = ['SepalaComp', 'SepalaLarg', 'PetalaComp', 'PetalaLarg']
iris_data[['SepalaComp', 'SepalaLarg','PetalaComp', 'PetalaLarg']].plot(figsize=(20,4))
```

```python
# Visualizando vários dados com plot
cols  = ['SepalaComp', 'SepalaLarg', 'PetalaComp', 'PetalaLarg']
iris_data[cols].plot(subplots=True, layout=(2,2), figsize=(20,7), grid=True, style='-o');
```

```python
# Visualizando um gráfico de barra
iris_data['Especie'].value_counts().plot.bar(figsize=(3,3), grid=True, 
                                             title='Quantidade por Especie');
```

```python
iris_data.columns
```

```python
# Visualizando um gráfico XY (Scatter Plot)
iris_data.plot.scatter(x='PetalaComp', y='PetalaLarg', title='Comp x Larg - Petala', grid=True, figsize=(4,4));
```

```python
# Visualizando um gráfico XY (Scatter Plot)
iris_data.plot.scatter(x='PetalaComp', y='PetalaLarg', title='Comp x Larg - Petala', grid=True, figsize=(4,4));
```

```python
iris_data['Especie'].value_counts().plot.pie(figsize=(4, 4));
```
