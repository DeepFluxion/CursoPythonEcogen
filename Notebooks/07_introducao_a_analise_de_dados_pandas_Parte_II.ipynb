## Vamos gerar estat√≠sticas para compreender os dados usando a fun√ß√£o `.describe()`

Assim teremos uma tabela onde para cada coluna:
- count: √© a contagem dos dados
- mean : √© a media
- std : √© o desvio padr√£o
- min : √© o valor m√≠nimo
- 25 %: √© o valor limite dos primeiros 25% dos dados (1o. quartil)
- 50 %: √© o valor limite dos primeiros 50% dos dados (2o. quartil)
- 75 %: √© o valor limite dos primeiros 75% dos dados (3o. quartil)
- max : √© o valor m√°ximo

```python
iris_data.describe()
```

```python
# Checando por missing values (valores ausentes)
iris_data.isnull().sum()
```

```python
# Checando por valores duplicados
duplicates = iris_data[iris_data.duplicated()]
print("Numero de duplicatas:", len(duplicates))
```
