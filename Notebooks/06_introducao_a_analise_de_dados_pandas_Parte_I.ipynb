````python
# a) cria_lista() uma fun√ß√£o que interaja com o usu√°rio para inserir uma lista de n√∫meros e retorne essa lista

def cria_lista():
    # Cria a lista 
    n = int(input("Quantos n√∫meros ter√° a lista: ")) # pergunte quantos n√∫eros ter√° a lista e armazene em n
    print('+---------------------------------------------------------------+')
    print(f"A Lista ter√° {n} n√∫meros")# informe a quantidade de n√∫meros da lista para o usu√°rios
    lista = [] #crie uma lista vazia
    for i in range(n): #Loop para repetir o n vezes a entrada da lista
        x = float(input(f"Qual o item {i} da lista: ")) #pergunte qual √© o item i da lista
        lista.append(x) #adicione o item x √† lista
    return lista
````

````python
# c) menor_numero() retorna o menor elemento
def menor_numero(lista): # Temos um par√¢metro aqui a lista de qual iremos selecionar o menor n√∫mero
    menor_numero = min(lista) # Armazenando o menor n√∫mero
    print('+---------------------------------------------------------------+')
    print(f'O menor valor √© {menor_numero}')
    return menor_numero # retorna o menor numero
````

````python
# d) numero_pares() retorna uma lista contendo n√∫meros pares
def numero_pares(lista): # Temos um par√¢metro aqui a lista de qual iremos selecionar os n√∫meros pares
    lista_pares = [] # Cria uma lista vazia para armazenar os numeros pares
    # Loop para percorrer a lista
    print('+---------------------------------------------------------------+')
    for item in lista:
    # verificando se resto da divis√£o com do item por 2 √© 0 com o operador m√≥dulo %
        e_par = item%2==0 
        if e_par: # se o resultado for verdadeiro executa estas fun√ß√µes
            print(f'o n√∫mero {item} √© um n√∫mero par') # imprimindo o item par
            lista_pares.append(item)
    return lista_pares       
````

# O que √© o desvio-padr√£o?

O desvio-padr√£o √© uma medida de dispers√£o do conjunto, ou seja, uma medida que indica qu√£o uniformes s√£o os dados do conjunto. O desvio-padr√£o demonstra a dist√¢ncia dos valores em rela√ß√£o √† m√©dia do conjunto, quanto mais pr√≥ximo de 0 for o desvio-padr√£o, menos dispersos s√£o os dados daquele conjunto. O desvio-padr√£o pode ser representado por $Dp$ ou pela letra grega $\sigma$ (sigma).

De modo geral, temos que:
- Quanto maior o desvio-padr√£o, **mais** dispersos s√£o os dados do conjunto (menos regular, menos homog√™neo).
- Quanto menor o desvio-padr√£o, **menos** dispersos s√£o os dados do conjunto (mais regular, mais homog√™neo).

## F√≥rmula do desvio-padr√£o
A f√≥rmula para calcular o desvio-padr√£o √©:

$$
\sigma = \sqrt(\frac{\sum_{i=0}^{n}(x_{i} - \bar{x})^2}{n})
$$

- $\sigma$: desvio-padr√£o

- $n$: quantidade de elementos no conjunto

- $x_i$: elementos do conjunto

- $\bar{x}$: m√©dia do conjunto

A f√≥rmula diz que o desvio-padr√£o √© a raiz quadrada da somat√≥ria da diferen√ßa entre cada um dos elementos do conjunto com a m√©dia, dividido pela quantidade de elementos do conjunto.

# Calculando o desvio padr√£o de uma lista

1. Para calcular a raiz quadrada de um n√∫mero o python j√° tem uma fun√ß√£o implementada chamada `sqrt()` para us√°-la temos de importar o m√≥dulo `math`, da seguinte forma:
```python
import math
num = float(input("Entre com um n√∫mero:\n"))
raiz = math.sqrt(num)
print(f'\nA raiz quadrada de {num} √© {raiz}\n')
```

2. Para calcular $(x_{i} - \bar{x})$ temos de calcular a m√©dia (usando nossa fun√ß√£o calcula_media(lista) e subtrair de cada elemento i da lista e elevar ao quadro, o que em Python seria:

```python
lista_dif = [] #cria uma lista vazia
media = calcula_media(lista) #calcula a media

for i in range(len(lista)): #percorre a lista gerando indices
    dif = lista[i]-media # calcula a diferen√ßa com a media
    lista_dif.append(dif**2) # calcula o quadrado da diferen√ßa e adiciona √† lista
```

3. Calcula a raiz quadrada com `math.sqrt()` da soma das diferen√ßas $\sqrt{\sum_{i=0}^{n}(x_{i} - \bar{x})^2}$ em python temos:
```python
soma_dif = calcula_soma(lista_dif) # somat√≥rio das diferen√ßas de lista_dif
raiz = math.sqrt(soma_dif) # Calcula raiz quadrada da soma de lista dif
sigma = raiz/n_elementos(lista) # divide raiz pelo numero de elementos de lista (ou de lista_dif)
```

# Colocando tudo junto numa fun√ß√£o

````python
import math

def calcula_desv_pad(lista):
    # Passo 1
    lista_dif = [] #cria uma lista vazia
    media = calcula_media(lista) #calcula a media
    
    for i in range(len(lista)): #percorre a lista gerando indices
        dif = lista[i]-media # calcula a diferen√ßa com a media
        lista_dif.append(dif**2) # calcula o quadrado da diferen√ßa e adiciona √† lista
    # Passo 2
    soma_dif = calcula_soma(lista_dif) # somat√≥rio das diferen√ßas de lista_dif
    raiz = math.sqrt(soma_dif) # Calcula raiz quadrada da soma de lista dif
    sigma = raiz/n_elementos(lista) # divide raiz pelo numero de elementos de lista (ou de lista_dif)
    print('+---------------------------------------------------------------+')
    print(f'o desvio padr√£o da lista √© : {sigma}')
    return sigma
````

````python
sigma = calcula_desv_pad(lista)
````
