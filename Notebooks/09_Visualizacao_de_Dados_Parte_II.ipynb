**Gráfico de Contagem:**
- O Gráfico de Contagem exibe as contagens de observações em cada categoria. Adequado para visualizar a frequência ou distribuição de dados categóricos.
- Use `sns.countplot(x, data=dataframe)` para criar um gráfico de contagem. Ao observar o comando, você pode ver por que eu disse que o Seaborn adora o pandas! Ou seja, ao contrário do Matplotlib, aqui você não precisa passar a série para o comando de plotagem ( `dataframe[column]` ), há um parâmetro chamado `data` para o qual você precisa atribuir seu dataframe, então para os parâmetros `x` e `y` você pode simplesmente atribuir o nome da sua coluna.

- Antes de vermos o gráfico de contagem, vamos importar um conjunto de dados disponível no Seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Carrega o "tips" dataset (gorjetas)
tips = sns.load_dataset("tips")

tips.head()
```

O conjunto de dados de **tips** trata das **tips** dadas pelas pessoas em um restaurante. Vamos ver um gráfico de contagem para a categoria de dia.

```python
# Count Plot
plt.figure(figsize=(5, 5))
sns.countplot(x="day", data=tips, hue='day', palette="Set1")
plt.title("Contagem de Gorjetas por Dia da Semana")
plt.show()
```

**Swarm Plot:**
- O Swarm Plot geralmente possui uma variável categórica no eixo x e uma variável numérica no eixo y, e exibe pontos de dados individuais ao longo de cada categoria. Isso nos fornece uma visualização da distribuição e densidade dos pontos de dados dentro das categorias.

**Principais características:**
- Minimiza a sobreposição entre os pontos de dados, o que significa que cada ponto de dados é plotado de forma a não se sobrepor a outros pontos na mesma categoria.
- Isso facilita a visualização da densidade de dados.
- Difícil de visualizar com conjuntos de dados maiores, sendo eficaz com conjuntos de dados relativamente pequenos.

**Uso:**
- Use `sns.swarmplot(x, y, data)` para fazer o swarm plot e use a paleta (palette) para alterar as cores.

```python
# Swarm Plot
plt.figure(figsize=(5, 5))
sns.swarmplot(x="day", y="total_bill", hue='day',data=tips, palette="viridis")
plt.title("Distribuição Total da Conta por Dia da Semana")
plt.xlabel("Dia da Semana")
plt.ylabel("Total Conta ($)")
plt.show()
```

**Gráfico de Pontos**    
O Gráfico de Pontos geralmente possui uma variável categórica no eixo x e uma variável numérica no eixo y e pode fornecer insights sobre a distribuição e tendência central dos dados dentro de cada categoria, como uma linha ou um ponto.

```python
# Point Plot
plt.figure(figsize=(5, 5))
sns.pointplot(x="day", y="total_bill", data=tips,errorbar="sd",hue='day', palette="pastel")
plt.title("Média do Total da Conta por Dia da Semana")
plt.xlabel("Dia da Semana")
plt.ylabel("Média do Total da Conta ($)")
plt.show()
```

- O ponto ou linha representa o valor médio ou mediano da variável numérica para cada categoria.
- Permite visualizar rapidamente se existem diferenças nos valores médios ou medianos entre as categorias.
- Os mesmos casos de uso se aplicam aqui, mas observaremos a tendência central por meio deste gráfico.
- Fornece uma ideia do que são os valores médios, baixos e altos em cada classe.

**Categorical Box Plot:**
- Um Box Plot Categórico, frequentemente referido simplesmente como um Box Plot ou Gráfico de Caixa e Bigodes, é um tipo de gráfico categórico usado para visualizar a distribuição de uma variável numérica dentro de diferentes categorias ou grupos.
- Ele exibe quartis (mediana, quartis superior e inferior), possíveis valores discrepantes e a dispersão geral dos dados.

**Uso:**
- Use `sns.boxplot(x, y, data)` para criar um gráfico de caixa categórico, mas este gráfico não é necessariamente para categorias, você pode até mesmo passar uma coluna numérica para o eixo x.

```python
# Categorical Box Plot
plt.figure(figsize=(5, 3))
sns.boxplot(x="time", y="total_bill", data=tips, hue='time',palette="coolwarm")
plt.title("Distribuição do Total da Conta por Horário da Refeição")
plt.xlabel("Horário da Refeição")
plt.ylabel("Conta Total ($)")
plt.show()
```

- O gráfico apresenta uma "caixa" retangular que representa a amplitude interquartil (IQR) dos dados, com uma linha no centro indicando a mediana dos dados.
- "Bigodes" se estendem a partir da caixa até os valores mínimo e máximo dentro de uma faixa definida, frequentemente 1,5 vezes o IQR.
- Pontos fora dos bigodes são considerados possíveis valores discrepantes.
- Útil para obter estatísticas básicas de uma variável categórica em relação a uma variável numérica.
- Também pode ser usado com duas variáveis numéricas, como tamanho e total de contas em conjuntos de dados de gorjetas.

**Categorical Violin Plot:**
- A variável categórica determina os grupos ou categorias, e uma variável numérica é usada para criar os Violin Plots dentro de cada categoria.
- A largura do gráfico em um ponto específico indica a densidade dos pontos de dados nesse valor, e a parte central de um Violin Plot se assemelha a um violino, daí o nome.
- Dentro do Violino, frequentemente há uma representação de gráfico de caixa que inclui a mediana, quartis e possíveis valores discrepantes.
- Isso fornece um resumo da tendência central e da dispersão dos dados para proporcionar uma visão abrangente da distribuição de uma variável numérica dentro de cada categoria.
- Use `sns.violinplot(x, y, data)` para criar um gráfico de violino.

```python
# Categorical Violin Plot
plt.figure(figsize=(5, 3))
sns.violinplot(x="day", y="total_bill", data=tips, hue='day',palette="Set2")
plt.title("Total Bill Distribution by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

**Violin Plot Categórico:**
- Combina características de um gráfico de caixa e um gráfico de densidade, observando a densidade no gráfico de enxame.
- Eficaz para comparar a distribuição de dados numéricos em diferentes categorias.
- Permite visualizar não apenas a tendência central e a dispersão, mas também a forma e a assimetria da distribuição.

**Cat Plot (Gráfico Cat):**
- O Cat Plot, abreviação de "Categorical Plot", é uma função de plotagem muito poderosa no Seaborn.
- Permite criar todos os gráficos categóricos mencionados anteriormente com apenas um parâmetro chamado 'kind' no comando `sns.catplot(x, y)`.
- Você pode usar `kind=bar`, `swarm`, `box`, `violin`, `count`, `point`, etc.
- Os Cat Plots funcionam bem para visualizar relações entre variáveis categóricas e variáveis numéricas.
- É como um canivete suíço para os gráficos categóricos.

**Gráficos Univariados:**
- Os gráficos univariados são usados para visualizar e analisar uma única variável por vez.
- São úteis para entender a distribuição e tendência central, identificar valores discrepantes e verificar padrões ou tendências dentro de uma única variável.

**Gráfico KDE:**
- O Gráfico KDE (Estimação de Densidade do Núcleo) é um tipo de visualização de dados usado para estimar a função de densidade de probabilidade de uma variável contínua.
- A variável contínua é representada no eixo x e o eixo y representa a área de densidade de probabilidade estimada sob a curva, com a soma igual a 1, indicando que a distribuição dos dados é normalizada.
- Use `sns.kdeplot(data=dataframe[column])` para criar um gráfico KDE para uma variável contínua.
- Você pode personalizar a aparência do Gráfico KDE ajustando parâmetros como largura de banda, tipo de kernel, cor e outros.

```python
# KDE Plot
plt.figure(figsize=(5, 3))
sns.kdeplot(data=tips['total_bill'], fill=True)
plt.title("Estimação de Densidade do Kernel (Núcleo) da Conta Total")
plt.xlabel("Total Conta ($)")
plt.ylabel("Densidad")
plt.show()
```

**Gráficos KDE:**
- Os gráficos KDE fornecem insights sobre a forma, picos, modas e dispersão dos dados.
- São valiosos para estimar a densidade de probabilidade dos pontos de dados, o que pode ser útil em análises estatísticas e testes de hipóteses.

**Rug Plot (Gráfico de Tapete):**
- Um Rug Plot consiste em linhas verticais (ou "ticks") posicionadas ao longo de um único eixo (geralmente o eixo x).
- Cada tick representa a localização de um ponto de dados individual, e a proximidade dos ticks representa a densidade dos pontos de dados; áreas mais densas têm mais ticks próximos.
- Use `sns.rugplot(data=dataframe[column])` para criar um rug plot.
- Você pode personalizar a aparência dos Rug Plots ajustando parâmetros como altura, cor e orientação dos ticks.

```python
# Rug Plot
plt.figure(figsize=(5, 3))
sns.rugplot(x=tips['total_bill'], height=0.5)
plt.title("Rug Plot da Conta Total")
plt.xlabel("Conta Total ($)")
plt.ylabel("Densidade")
plt.show()
```

**Gráfico de Tapete:**

- A partir do KDE, observamos que o pico do valor da conta está em algum lugar entre 10–20, mas através do Rug plot, temos uma ideia de que a densidade é maior em 15–20.
- É particularmente útil quando se deseja ver a posição exata dos pontos de dados individuais.
- Os Rug Plots são frequentemente combinados com histogramas, gráficos KDE ou gráficos de caixa para fornecer contexto e detalhes adicionais.
- Podem ajudar a destacar valores discrepantes ou áreas de alta densidade de dados.

** Histogramas com histplot**
Uma alternativa é misturar histograms com a função kde, você pode usar o `sns.histplot(data=dataframe[column])` e as opções `kde=True` e `stat = density`

```python
plt.figure(figsize=(5, 3))
sns.histplot(tips["total_bill"], kde=True, stat="density")
plt.title("Hist Plot com densidade da Conta Total")
plt.xlabel("Conta Total ($)")
plt.ylabel("Densidade")
plt.show()
```

**Gráficos Bivariados:**

- Envolve a visualização e análise da relação entre duas variáveis simultaneamente.
- São utilizados para explorar como duas variáveis estão relacionadas ou correlacionadas.
- Gráficos comuns com o Matplotlib incluem `sns.scatterplot(x, y, data)` e `sns.lineplot(x, y, data)` para gráficos de dispersão e de linha.

Gráfico de Regressão

- Um Gráfico de Regressão foca na relação entre duas variáveis numéricas: a variável independente (frequentemente no eixo x) e a variável dependente (no eixo y).
- Os pontos de dados individuais são exibidos como pontos e o elemento central de um Gráfico de Regressão é a linha ou curva de regressão, que representa o modelo matemático de melhor ajuste que descreve a relação entre as variáveis.
- Use `sns.regplot(x, y, data)` para criar um gráfico de regressão.

```python
# Regression Plot
plt.figure(figsize=(5, 3))
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title("Regression Plot da Conta Total vs. Gorjeta")
plt.xlabel("Conta Total ($)")
plt.ylabel("Gorjeta ($)")
plt.show()
```

- A linha de regressão representa o melhor modelo linear para prever gorjetas com base nos totais da conta.
- Os pontos de dispersão mostram dados individuais.
- Permite observar como os pontos de dados se agrupam em torno da linha de regressão.
- Útil para compreender a relação linear entre as variáveis.

**Joint Plot (Gráfico Conjunto):**

- Combina gráficos de dispersão, histogramas e gráficos de densidade para visualizar a relação entre duas variáveis numéricas.
- O elemento central de um Joint Plot é um Gráfico de Dispersão que exibe os pontos de dados das duas variáveis em relação uma à outra.
- Ao longo dos eixos x e y do Gráfico de Dispersão, há histogramas ou gráficos de Estimação de Densidade do Núcleo (KDE) para cada variável individual.
- Esses gráficos marginais mostram a distribuição de cada variável separadamente.
- Use `sns.jointplot(x, y, data=dataframe, kind)`, onde `kind` pode ser um dos seguintes: ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid'].

```python
# Joint Plot
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter");
plt.show();
```

**Hexbin Plot (Gráfico de Binagem Hexagonal):**

- Um Hexbin plot, abreviação para Hexagonal Binning plot, agrupa pontos de dados em bins hexagonais, permitindo visualizar a densidade de dados e padrões de forma mais eficaz.
- São especialmente valiosos com conjuntos de dados grandes, quando os gráficos de dispersão com pontos individuais se tornam muito lotados e difíceis de interpretar.
- Você pode criar um Hexbin plot usando o parâmetro `kind` do gráfico conjunto.
- É possível personalizar o mapa de cores, o tamanho da grade e outros parâmetros do gráfico para ajustar a aparência do Hexbin Plot.

```python
# Hexbin Plot
plt.figure(figsize=(4, 4))
sns.jointplot(x="total_bill", y="tip",kind='hex', data=tips, gridsize=15, cmap="Blues")
plt.xlabel("Conta Total ($)")
plt.ylabel("Gorjeta ($)")
plt.show()
```

**Usando Parâmetros para Adicionar Dimensões:**

- **Usando o parâmetro Hue:** O uso do parâmetro "hue" adicionará cor ao gráfico com base na variável categórica fornecida, especificando uma cor única para cada uma das categorias. Este parâmetro pode ser usado em quase todos os tipos de gráficos, como `.scatterplot()`, `.boxplot()`, `.violinplot()`, `.lineplot()`, etc.

```python
# Violin Plot com Hue
plt.figure(figsize=(20, 5))
sns.violinplot(
    x="day",         # x-axis: Dias da Seman (categorica)
    y="total_bill",  # y-axis: Valor Total da Conta (numerico)
    data=tips,
    hue="sex",       # Cor por Gêneros (categorica)
    palette="Set1",  # paltea de Cores
    split=True       # Separe os violinos pela categoria hue
)

plt.title("Violin Plot com Hue para Conta Total por Dia e Genero")
plt.xlabel("Dia da Semana")
plt.ylabel("Conta Total ($)")
plt.legend(title="Genero")
plt.show()
```

Usamos a coluna "sex" para colorir os violinos com base no gênero, adicionando uma dimensão ao gráfico. Cada gênero é representado por uma cor diferente.

Além disso, o parâmetro "size" pode ser utilizado para adicionar outra dimensão numérica.

```python
tips.head()
```

```python
# Scatter Plot com Hue e Size
plt.figure(figsize=(10, 4))
sns.scatterplot(
    x="total_bill",
    y="tip",
    data=tips,
    hue="day",      # Color por dia (categorical)
    size="size",    # Varia o tamanho do marker size pela coluna size (numerica)
    sizes=(20, 200),  # Define a faixa de size for markers
    palette="Set1"  # paleta de cores
)
plt.title("Scatter Plot com Hue e Size para Dataset Tips")
plt.xlabel("Conta Total ($)")
plt.ylabel("Gorjeta ($)")
plt.legend(title="Dia")
plt.show()
```

**Gráfico Relacional:**

- Permite visualizar a relação entre duas variáveis numéricas.
- Inclui dimensões adicionais categóricas ou numéricas.
- Uso do `sns.relplot(x, y, data, hue, size, style)`.
- O parâmetro "hue" colore os pontos de dados por uma variável categórica.
- O parâmetro "size" varia o tamanho do marcador com base em uma variável numérica.
- O parâmetro "style" diferencia marcadores ou linhas por uma variável categórica.
- O parâmetro "kind" especifica o tipo de gráfico relacional desejado.

```python
# Create a facet grid using a Relational Plot (relplot)
g = sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    hue="time",
    col="day",  # Separa gráfico por dia (colunas)
    row="sex",  # Separa gráfico por gênero  (linhas)
    palette="Set1",
    height=2.5,  # Altura de cada subgráfico
    aspect=1.2  # Aspect ratio de cada subplot
)

# Set titles and labels for the facets
g.set_titles(col_template="Dia {col_name}", row_template="Genero {row_name}")
g.set_axis_labels("Conta Total ($)", "Gorjeta ($)")
plt.suptitle("Relational Scatter Plots por Dia e Genero")
plt.subplots_adjust(top=0.9)  # Ajusta a posição do titulo
plt.show()
```

**Facet Grid:**

- Um Facet Grid é uma característica do Seaborn que permite criar uma grade de subplots, cada um representando um subconjunto diferente dos seus dados.
- São usados para comparar padrões ou relações com várias variáveis dentro de diferentes categorias.
- Use `sns.FacetGrid(data, col, row)` para criar um facet grid, que retorna o objeto de grade.
- Após criar o objeto da grade, você precisa mapeá-lo para qualquer gráfico de sua escolha.

```python
# Cria um Facet Grid de histogramas para dias diferentes
g = sns.FacetGrid(tips, col="day", height=4, aspect=1.2)
g.map(sns.histplot, "total_bill", kde=True)
g.set_axis_labels("Conta Total Bill ($)", "Frequencia")
g.set_titles(col_template="Dia {col_name}")
#plt.savefig('facet_grid_hist_plot.png')
plt.show()
```

**Pair Plot:**

- Um Pair plot fornece uma grade de gráficos de dispersão e histogramas, onde cada gráfico mostra a relação entre duas variáveis, razão pela qual também é chamado de Pairwise Plot ou Scatterplot Matrix.
- As células diagonais geralmente exibem histogramas ou gráficos de densidade de kernel para variáveis individuais, mostrando suas distribuições.
- As células fora da diagonal na grade frequentemente exibem gráficos de dispersão, mostrando como duas variáveis estão relacionadas.
- Os Pair Plots são particularmente úteis para entender padrões, correlações e distribuições em várias dimensões nos seus dados.
- Use `sns.pairplot(data)` para criar um pairplot. Você pode personalizar a aparência dos Pair Plots, como mudar o tipo de gráficos (dispersão, KDE, etc.), cores, marcadores e mais.
- Se você deseja alterar os gráficos diagonais, pode fazer isso usando o parâmetro `diag_kind`.

```python
# carrega o "iris" dataset
iris = sns.load_dataset("iris")

# Pair Plot
sns.set(style="ticks")
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.show()
```

**Pair Grid:**

- Usando um pair grid, você pode personalizar os gráficos inferiores, superiores e diagonais individualmente.

```python
# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Create a Facet Grid of pairwise scatterplots
g = sns.PairGrid(iris, hue="species")
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot, kde_kws={"color": "k"})
g.map_lower(sns.kdeplot)
g.add_legend()
plt.show()
```

**Matrix Plots:**

- Esses gráficos visualizam relacionamentos dentro de matrizes ou grades de dados.

**Heatmap:**

- Comecemos com o gráfico mais comumente usado, o Heatmap. Um Heatmap fornecerá uma ideia de como uma variável está correlacionada com todas as outras variáveis (numéricas) no quadro de dados.
- Precisamos passar a matriz de correlação para o heatmap e ele plotará um gráfico com diferentes cores, que representarão a correlação.
- Use o método `dataframe.corr()` do pandas para obter as correlações e, em seguida, passe para o método `sns.heatmap(corr_matrix)` para plotar o heatmap.
- Se quiser saber mais sobre correlações viste este [link](https://pt.wikipedia.org/wiki/Correla%C3%A7%C3%A3o)

```python
# Importe o Seaborn
import seaborn as sns

# Para ver os datasets oferecidos pelo seaborn
print(sns.get_dataset_names())
# vamos usar um dataset oferecidos seaborn
data = sns.load_dataset("flights")
pivot_data = pd.pivot_table(data, values='passengers', index='month', columns='year')
data['passengers'].plot(figsize=(20,3))
```

```python
pivot_data
```

```python
plt.figure(figsize=(7,7))
# Creia um heatmap da tabela pivot
sns.heatmap(pivot_data, annot=True, cmap="coolwarm", fmt='g')
plt.title("Mapa de Calor do Dataset Fligths")
plt.show()
```

**Cluster Map:**

O Cluster Map é uma ferramenta valiosa para revelar clusters ou grupos de linhas e colunas semelhantes em um conjunto de dados.

- É uma variação do heatmap do Seaborn que não apenas exibe os dados como uma grade de células coloridas, mas também organiza linhas e colunas de forma a agrupar dados semelhantes.
- Utiliza algoritmos de agrupamento hierárquico para reordenar linhas e colunas no heatmap.
- Além disso, inclui dendrogramas ao lado do heatmap agrupados dendrogramas são diagramas em forma de árvore que mostram as relações hierárquicas entre linhas e colunas.

Para criar um Cluster Map, utilize `sns.clustermap(data)`.

```python
plt.figure(figsize=(7,7))
sns.clustermap(pivot_data, annot=True, cmap="coolwarm", fmt='g')
```
