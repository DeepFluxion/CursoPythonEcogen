**Gr√°fico de Contagem:**
- O Gr√°fico de Contagem exibe as contagens de observa√ß√µes em cada categoria. Adequado para visualizar a frequ√™ncia ou distribui√ß√£o de dados categ√≥ricos.
- Use `sns.countplot(x, data=dataframe)` para criar um gr√°fico de contagem. Ao observar o comando, voc√™ pode ver por que eu disse que o Seaborn adora o pandas! Ou seja, ao contr√°rio do Matplotlib, aqui voc√™ n√£o precisa passar a s√©rie para o comando de plotagem ( `dataframe[column]` ), h√° um par√¢metro chamado `data` para o qual voc√™ precisa atribuir seu dataframe, ent√£o para os par√¢metros `x` e `y` voc√™ pode simplesmente atribuir o nome da sua coluna.

- Antes de vermos o gr√°fico de contagem, vamos importar um conjunto de dados dispon√≠vel no Seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Carrega o "tips" dataset (gorjetas)
tips = sns.load_dataset("tips")

tips.head()
```

O conjunto de dados de **tips** trata das **tips** dadas pelas pessoas em um restaurante. Vamos ver um gr√°fico de contagem para a categoria de dia.

```python
# Count Plot
plt.figure(figsize=(5, 5))
sns.countplot(x="day", data=tips, hue='day', palette="Set1")
plt.title("Contagem de Gorjetas por Dia da Semana")
plt.show()
```

**Swarm Plot:**
- O Swarm Plot geralmente possui uma vari√°vel categ√≥rica no eixo x e uma vari√°vel num√©rica no eixo y, e exibe pontos de dados individuais ao longo de cada categoria. Isso nos fornece uma visualiza√ß√£o da distribui√ß√£o e densidade dos pontos de dados dentro das categorias.

**Principais caracter√≠sticas:**
- Minimiza a sobreposi√ß√£o entre os pontos de dados, o que significa que cada ponto de dados √© plotado de forma a n√£o se sobrepor a outros pontos na mesma categoria.
- Isso facilita a visualiza√ß√£o da densidade de dados.
- Dif√≠cil de visualizar com conjuntos de dados maiores, sendo eficaz com conjuntos de dados relativamente pequenos.

**Uso:**
- Use `sns.swarmplot(x, y, data)` para fazer o swarm plot e use a paleta (palette) para alterar as cores.

```python
# Swarm Plot
plt.figure(figsize=(5, 5))
sns.swarmplot(x="day", y="total_bill", hue='day',data=tips, palette="viridis")
plt.title("Distribui√ß√£o Total da Conta por Dia da Semana")
plt.xlabel("Dia da Semana")
plt.ylabel("Total Conta ($)")
plt.show()
```

**Gr√°fico de Pontos**    
O Gr√°fico de Pontos geralmente possui uma vari√°vel categ√≥rica no eixo x e uma vari√°vel num√©rica no eixo y e pode fornecer insights sobre a distribui√ß√£o e tend√™ncia central dos dados dentro de cada categoria, como uma linha ou um ponto.

```python
# Point Plot
plt.figure(figsize=(5, 5))
sns.pointplot(x="day", y="total_bill", data=tips,errorbar="sd",hue='day', palette="pastel")
plt.title("M√©dia do Total da Conta por Dia da Semana")
plt.xlabel("Dia da Semana")
plt.ylabel("M√©dia do Total da Conta ($)")
plt.show()
```

- O ponto ou linha representa o valor m√©dio ou mediano da vari√°vel num√©rica para cada categoria.
- Permite visualizar rapidamente se existem diferen√ßas nos valores m√©dios ou medianos entre as categorias.
- Os mesmos casos de uso se aplicam aqui, mas observaremos a tend√™ncia central por meio deste gr√°fico.
- Fornece uma ideia do que s√£o os valores m√©dios, baixos e altos em cada classe.

**Categorical Box Plot:**
- Um Box Plot Categ√≥rico, frequentemente referido simplesmente como um Box Plot ou Gr√°fico de Caixa e Bigodes, √© um tipo de gr√°fico categ√≥rico usado para visualizar a distribui√ß√£o de uma vari√°vel num√©rica dentro de diferentes categorias ou grupos.
- Ele exibe quartis (mediana, quartis superior e inferior), poss√≠veis valores discrepantes e a dispers√£o geral dos dados.

**Uso:**
- Use `sns.boxplot(x, y, data)` para criar um gr√°fico de caixa categ√≥rico, mas este gr√°fico n√£o √© necessariamente para categorias, voc√™ pode at√© mesmo passar uma coluna num√©rica para o eixo x.

```python
# Categorical Box Plot
plt.figure(figsize=(5, 3))
sns.boxplot(x="time", y="total_bill", data=tips, hue='time',palette="coolwarm")
plt.title("Distribui√ß√£o do Total da Conta por Hor√°rio da Refei√ß√£o")
plt.xlabel("Hor√°rio da Refei√ß√£o")
plt.ylabel("Conta Total ($)")
plt.show()
```

- O gr√°fico apresenta uma "caixa" retangular que representa a amplitude interquartil (IQR) dos dados, com uma linha no centro indicando a mediana dos dados.
- "Bigodes" se estendem a partir da caixa at√© os valores m√≠nimo e m√°ximo dentro de uma faixa definida, frequentemente 1,5 vezes o IQR.
- Pontos fora dos bigodes s√£o considerados poss√≠veis valores discrepantes.
- √ötil para obter estat√≠sticas b√°sicas de uma vari√°vel categ√≥rica em rela√ß√£o a uma vari√°vel num√©rica.
- Tamb√©m pode ser usado com duas vari√°veis num√©ricas, como tamanho e total de contas em conjuntos de dados de gorjetas.

**Categorical Violin Plot:**
- A vari√°vel categ√≥rica determina os grupos ou categorias, e uma vari√°vel num√©rica √© usada para criar os Violin Plots dentro de cada categoria.
- A largura do gr√°fico em um ponto espec√≠fico indica a densidade dos pontos de dados nesse valor, e a parte central de um Violin Plot se assemelha a um violino, da√≠ o nome.
- Dentro do Violino, frequentemente h√° uma representa√ß√£o de gr√°fico de caixa que inclui a mediana, quartis e poss√≠veis valores discrepantes.
- Isso fornece um resumo da tend√™ncia central e da dispers√£o dos dados para proporcionar uma vis√£o abrangente da distribui√ß√£o de uma vari√°vel num√©rica dentro de cada categoria.
- Use `sns.violinplot(x, y, data)` para criar um gr√°fico de violino.

```python
# Categorical Violin Plot
plt.figure(figsize=(5, 3))
sns.violinplot(x="day", y="total_bill", data=tips, hue='day',palette="Set2")
plt.title("Total Bill Distribution by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

**Violin Plot Categ√≥rico:**
- Combina caracter√≠sticas de um gr√°fico de caixa e um gr√°fico de densidade, observando a densidade no gr√°fico de enxame.
- Eficaz para comparar a distribui√ß√£o de dados num√©ricos em diferentes categorias.
- Permite visualizar n√£o apenas a tend√™ncia central e a dispers√£o, mas tamb√©m a forma e a assimetria da distribui√ß√£o.

**Cat Plot (Gr√°fico Cat):**
- O Cat Plot, abrevia√ß√£o de "Categorical Plot", √© uma fun√ß√£o de plotagem muito poderosa no Seaborn.
- Permite criar todos os gr√°ficos categ√≥ricos mencionados anteriormente com apenas um par√¢metro chamado 'kind' no comando `sns.catplot(x, y)`.
- Voc√™ pode usar `kind=bar`, `swarm`, `box`, `violin`, `count`, `point`, etc.
- Os Cat Plots funcionam bem para visualizar rela√ß√µes entre vari√°veis categ√≥ricas e vari√°veis num√©ricas.
- √â como um canivete su√≠√ßo para os gr√°ficos categ√≥ricos.

**Gr√°ficos Univariados:**
- Os gr√°ficos univariados s√£o usados para visualizar e analisar uma √∫nica vari√°vel por vez.
- S√£o √∫teis para entender a distribui√ß√£o e tend√™ncia central, identificar valores discrepantes e verificar padr√µes ou tend√™ncias dentro de uma √∫nica vari√°vel.

**Gr√°fico KDE:**
- O Gr√°fico KDE (Estima√ß√£o de Densidade do N√∫cleo) √© um tipo de visualiza√ß√£o de dados usado para estimar a fun√ß√£o de densidade de probabilidade de uma vari√°vel cont√≠nua.
- A vari√°vel cont√≠nua √© representada no eixo x e o eixo y representa a √°rea de densidade de probabilidade estimada sob a curva, com a soma igual a 1, indicando que a distribui√ß√£o dos dados √© normalizada.
- Use `sns.kdeplot(data=dataframe[column])` para criar um gr√°fico KDE para uma vari√°vel cont√≠nua.
- Voc√™ pode personalizar a apar√™ncia do Gr√°fico KDE ajustando par√¢metros como largura de banda, tipo de kernel, cor e outros.

```python
# KDE Plot
plt.figure(figsize=(5, 3))
sns.kdeplot(data=tips['total_bill'], fill=True)
plt.title("Estima√ß√£o de Densidade do Kernel (N√∫cleo) da Conta Total")
plt.xlabel("Total Conta ($)")
plt.ylabel("Densidad")
plt.show()
```

**Gr√°ficos KDE:**
- Os gr√°ficos KDE fornecem insights sobre a forma, picos, modas e dispers√£o dos dados.
- S√£o valiosos para estimar a densidade de probabilidade dos pontos de dados, o que pode ser √∫til em an√°lises estat√≠sticas e testes de hip√≥teses.

**Rug Plot (Gr√°fico de Tapete):**
- Um Rug Plot consiste em linhas verticais (ou "ticks") posicionadas ao longo de um √∫nico eixo (geralmente o eixo x).
- Cada tick representa a localiza√ß√£o de um ponto de dados individual, e a proximidade dos ticks representa a densidade dos pontos de dados; √°reas mais densas t√™m mais ticks pr√≥ximos.
- Use `sns.rugplot(data=dataframe[column])` para criar um rug plot.
- Voc√™ pode personalizar a apar√™ncia dos Rug Plots ajustando par√¢metros como altura, cor e orienta√ß√£o dos ticks.

```python
# Rug Plot
plt.figure(figsize=(5, 3))
sns.rugplot(x=tips['total_bill'], height=0.5)
plt.title("Rug Plot da Conta Total")
plt.xlabel("Conta Total ($)")
plt.ylabel("Densidade")
plt.show()
```

**Gr√°fico de Tapete:**

- A partir do KDE, observamos que o pico do valor da conta est√° em algum lugar entre 10‚Äì20, mas atrav√©s do Rug plot, temos uma ideia de que a densidade √© maior em 15‚Äì20.
- √â particularmente √∫til quando se deseja ver a posi√ß√£o exata dos pontos de dados individuais.
- Os Rug Plots s√£o frequentemente combinados com histogramas, gr√°ficos KDE ou gr√°ficos de caixa para fornecer contexto e detalhes adicionais.
- Podem ajudar a destacar valores discrepantes ou √°reas de alta densidade de dados.

** Histogramas com histplot**
Uma alternativa √© misturar histograms com a fun√ß√£o kde, voc√™ pode usar o `sns.histplot(data=dataframe[column])` e as op√ß√µes `kde=True` e `stat = density`

```python
plt.figure(figsize=(5, 3))
sns.histplot(tips["total_bill"], kde=True, stat="density")
plt.title("Hist Plot com densidade da Conta Total")
plt.xlabel("Conta Total ($)")
plt.ylabel("Densidade")
plt.show()
```

**Gr√°ficos Bivariados:**

- Envolve a visualiza√ß√£o e an√°lise da rela√ß√£o entre duas vari√°veis simultaneamente.
- S√£o utilizados para explorar como duas vari√°veis est√£o relacionadas ou correlacionadas.
- Gr√°ficos comuns com o Matplotlib incluem `sns.scatterplot(x, y, data)` e `sns.lineplot(x, y, data)` para gr√°ficos de dispers√£o e de linha.
- Veremos mais sobre alguns gr√°ficos menos comuns aqui.

Gr√°fico de Regress√£o

- Um Gr√°fico de Regress√£o foca na rela√ß√£o entre duas vari√°veis num√©ricas: a vari√°vel independente (frequentemente no eixo x) e a vari√°vel dependente (no eixo y).
- Os pontos de dados individuais s√£o exibidos como pontos e o elemento central de um Gr√°fico de Regress√£o √© a linha ou curva de regress√£o, que representa o modelo matem√°tico de melhor ajuste que descreve a rela√ß√£o entre as vari√°veis.
- Use `sns.regplot(x, y, data)` para criar um gr√°fico de regress√£o.

```python
# Regression Plot
plt.figure(figsize=(5, 3))
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title("Regression Plot da Conta Total vs. Gorjeta")
plt.xlabel("Conta Total ($)")
plt.ylabel("Gorjeta ($)")
plt.show()
```

- A linha de regress√£o representa o melhor modelo linear para prever gorjetas com base nos totais da conta.
- Os pontos de dispers√£o mostram dados individuais.
- Permite observar como os pontos de dados se agrupam em torno da linha de regress√£o.
- √ötil para compreender a rela√ß√£o linear entre as vari√°veis.

**Joint Plot (Gr√°fico Conjunto):**

- Combina gr√°ficos de dispers√£o, histogramas e gr√°ficos de densidade para visualizar a rela√ß√£o entre duas vari√°veis num√©ricas.
- O elemento central de um Joint Plot √© um Gr√°fico de Dispers√£o que exibe os pontos de dados das duas vari√°veis em rela√ß√£o uma √† outra.
- Ao longo dos eixos x e y do Gr√°fico de Dispers√£o, h√° histogramas ou gr√°ficos de Estima√ß√£o de Densidade do N√∫cleo (KDE) para cada vari√°vel individual.
- Esses gr√°ficos marginais mostram a distribui√ß√£o de cada vari√°vel separadamente.
- Use `sns.jointplot(x, y, data=dataframe, kind)`, onde `kind` pode ser um dos seguintes: ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid'].

```python
# Joint Plot
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter");
plt.show();
```

**Hexbin Plot (Gr√°fico de Binagem Hexagonal):**

- Um Hexbin plot, abrevia√ß√£o para Hexagonal Binning plot, agrupa pontos de dados em bins hexagonais, permitindo visualizar a densidade de dados e padr√µes de forma mais eficaz.
- S√£o especialmente valiosos com conjuntos de dados grandes, quando os gr√°ficos de dispers√£o com pontos individuais se tornam muito lotados e dif√≠ceis de interpretar.
- Voc√™ pode criar um Hexbin plot usando o par√¢metro `kind` do gr√°fico conjunto.
- √â poss√≠vel personalizar o mapa de cores, o tamanho da grade e outros par√¢metros do gr√°fico para ajustar a apar√™ncia do Hexbin Plot.

```python
# Hexbin Plot
plt.figure(figsize=(4, 4))
sns.jointplot(x="total_bill", y="tip",kind='hex', data=tips, gridsize=15, cmap="Blues")
plt.xlabel("Conta Total ($)")
plt.ylabel("Gorjeta ($)")
plt.show()
```

**Usando Par√¢metros para Adicionar Dimens√µes:**

- **Usando o par√¢metro Hue:** O uso do par√¢metro "hue" adicionar√° cor ao gr√°fico com base na vari√°vel categ√≥rica fornecida, especificando uma cor √∫nica para cada uma das categorias. Este par√¢metro pode ser usado em quase todos os tipos de gr√°ficos, como `.scatterplot()`, `.boxplot()`, `.violinplot()`, `.lineplot()`, etc.

```python
# Violin Plot com Hue
plt.figure(figsize=(20, 5))
sns.violinplot(
    x="day",         # x-axis: Dias da Seman (categorica)
    y="total_bill",  # y-axis: Valor Total da Conta (numerico)
    data=tips,
    hue="sex",       # Cor por G√™neros (categorica)
    palette="Set1",  # paltea de Cores
    split=True       # Separe os violinos pela categoria hue
)

plt.title("Violin Plot com Hue para Conta Total por Dia e Genero")
plt.xlabel("Dia da Semana")
plt.ylabel("Conta Total ($)")
plt.legend(title="Genero")
plt.show()
```

Usamos a coluna "sex" para colorir os violinos com base no g√™nero, adicionando uma dimens√£o ao gr√°fico. Cada g√™nero √© representado por uma cor diferente.

Al√©m disso, o par√¢metro "size" pode ser utilizado para adicionar outra dimens√£o num√©rica.

**Gr√°fico Relacional:**

- Permite visualizar a rela√ß√£o entre duas vari√°veis num√©ricas.
- Inclui dimens√µes adicionais categ√≥ricas ou num√©ricas.
- Uso do `sns.relplot(x, y, data, hue, size, style)`.
- O par√¢metro "hue" colore os pontos de dados por uma vari√°vel categ√≥rica.
- O par√¢metro "size" varia o tamanho do marcador com base em uma vari√°vel num√©rica.
- O par√¢metro "style" diferencia marcadores ou linhas por uma vari√°vel categ√≥rica.
- O par√¢metro "kind" especifica o tipo de gr√°fico relacional desejado.

```python
# Create a facet grid using a Relational Plot (relplot)
g = sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    hue="time",
    col="day",  # Separa gr√°fico por dia (colunas)
    row="sex",  # Separa gr√°fico por g√™nero  (linhas)
    palette="Set1",
    height=2.5,  # Altura de cada subgr√°fico
    aspect=1.2  # Aspect ratio de cada subplot
)

# Set titles and labels for the facets
g.set_titles(col_template="Dia {col_name}", row_template="Genero {row_name}")
g.set_axis_labels("Conta Total ($)", "Gorjeta ($)")
plt.suptitle("Relational Scatter Plots por Dia e Genero")
plt.subplots_adjust(top=0.9)  # Ajusta a posi√ß√£o do titulo
plt.show()
```

**Facet Grid:**

- Um Facet Grid √© uma caracter√≠stica do Seaborn que permite criar uma grade de subplots, cada um representando um subconjunto diferente dos seus dados.
- S√£o usados para comparar padr√µes ou rela√ß√µes com v√°rias vari√°veis dentro de diferentes categorias.
- Use `sns.FacetGrid(data, col, row)` para criar um facet grid, que retorna o objeto de grade.
- Ap√≥s criar o objeto da grade, voc√™ precisa mape√°-lo para qualquer gr√°fico de sua escolha.

```python
# Cria um Facet Grid de histogramas para dias diferentes
g = sns.FacetGrid(tips, col="day", height=4, aspect=1.2)
g.map(sns.histplot, "total_bill", kde=True)
g.set_axis_labels("Conta Total Bill ($)", "Frequencia")
g.set_titles(col_template="Dia {col_name}")
#plt.savefig('facet_grid_hist_plot.png')
plt.show()
```

**Pair Plot:**

- Um Pair plot fornece uma grade de gr√°ficos de dispers√£o e histogramas, onde cada gr√°fico mostra a rela√ß√£o entre duas vari√°veis, raz√£o pela qual tamb√©m √© chamado de Pairwise Plot ou Scatterplot Matrix.
- As c√©lulas diagonais geralmente exibem histogramas ou gr√°ficos de densidade de kernel para vari√°veis individuais, mostrando suas distribui√ß√µes.
- As c√©lulas fora da diagonal na grade frequentemente exibem gr√°ficos de dispers√£o, mostrando como duas vari√°veis est√£o relacionadas.
- Os Pair Plots s√£o particularmente √∫teis para entender padr√µes, correla√ß√µes e distribui√ß√µes em v√°rias dimens√µes nos seus dados.
- Use `sns.pairplot(data)` para criar um pairplot. Voc√™ pode personalizar a apar√™ncia dos Pair Plots, como mudar o tipo de gr√°ficos (dispers√£o, KDE, etc.), cores, marcadores e mais.
- Se voc√™ deseja alterar os gr√°ficos diagonais, pode fazer isso usando o par√¢metro `diag_kind`.

```python
# carrega o "iris" dataset
iris = sns.load_dataset("iris")

# Pair Plot
sns.set(style="ticks")
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.show()
```

**Pair Grid:**

- Usando um pair grid, voc√™ pode personalizar os gr√°ficos inferiores, superiores e diagonais individualmente.

```python
# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Create a Facet Grid of pairwise scatterplots
g = sns.PairGrid(iris, hue="species")
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot, kde_kws={"color": "k"})
g.map_lower(sns.kdeplot)
g.add_legend()
plt.show()
```
