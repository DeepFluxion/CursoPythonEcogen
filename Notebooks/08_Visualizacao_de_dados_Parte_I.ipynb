## Mas o que isso quer dizer:

O que é o conjunto de dados da flor Iris? É um dos conjuntos de dados mais conhecidos no mundo do aprendizado de máquina e da ciência de dados, e por um bom motivo. É composto por 150 registros de flores de íris, incluindo informações sobre o comprimento e largura das sépalas e pétalas, bem como o tipo de flor de íris. Nesta postagem do blog, exploraremos o conjunto de dados Iris e aprenderemos sobre as diferentes técnicas e métodos que podemos usar para analisá-lo e compreendê-lo. Seja você um iniciante ou um cientista de dados experiente, esta postagem fornecerá insights e dicas valiosas para trabalhar com esse conjunto de dados clássico.

O conjunto de dados foi introduzido pelo estatístico e biólogo britânico Ronald Fisher em seu artigo de 1936, “O uso de múltiplas medições em problemas taxonômicos.

## Vamos gerar estatísticas para compreender os dados usando a função `.describe()`

Assim teremos uma tabela onde para cada coluna:
- count: é a contagem dos dados
- mean : é a media
- std : é o desvio padrão
- min : é o valor mínimo
- 25 %: é o valor limite dos primeiros 25% dos dados (1o. quartil)
- 50 %: é o valor limite dos primeiros 50% dos dados (2o. quartil)
- 75 %: é o valor limite dos primeiros 75% dos dados (3o. quartil)
- max : é o valor máximo

# Exemplos de Gráficos com o Matplotlib

```python
from matplotlib import pyplot as plt # Importando o Matplotlib
import numpy as np # Importando o numpy para nos ajusar a criar os dados
```

## Um exemplo simples
Matplotlib representa graficamente seus dados em figuras (por exemplo, janelas, widgets Jupyter, etc.), cada uma das quais pode conter um ou mais eixos (axis), uma área onde os pontos podem ser especificados em termos de coordenadas x-y (ou theta-r em um gráfico polar, x-y-z em um gráfico 3D, etc.). A maneira mais simples de criar uma figura com eixos é usando pyplot.subplots. Podemos então usar Axes.plot para desenhar alguns dados nos eixos:

```python
fig, ax = plt.subplots()  # Cria uma figura contendo um single eixo (axes).
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plota alguns dado no eixo.
```

# Gráfico do Tipo Plot (Linhas)

```python
# Selcionando um estilo
plt.style.use('_mpl-gallery')

# crie os dada
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

# plot
fig, ax = plt.subplots(figsize=(4,4))
ax.plot(x, y, linewidth=3.0)

ax.set(xlim=(0, 10), xticks=np.arange(1, 10),
       ylim=(0, 10), yticks=np.arange(1, 10))

plt.show()
```

```python
 plt.style.available
```

```python
# Selcionando um estilo
plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots(figsize=(4,4))

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
# gráfico de barras
plt.style.use('_mpl-gallery')

# make data:
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = np.linspace(0, 8, 16)
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 0.5, len(x))

# plot
fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, alpha=.5, linewidth=0)
ax.plot(x, (y1 + y2)/2, linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

# plot:
fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.show()
```

```python
plt.style.use('_mpl-gallery')

# make data
x = np.arange(0, 10, 2)
ay = [1, 1.25, 2, 2.75, 3]
by = [1, 1, 1, 1, 1]
cy = [2, 1, 2, 1, 2]
y = np.vstack([ay, by, cy])

# plot
fig, ax = plt.subplots()

ax.stackplot(x, y)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
# Personalizando o Gráfcio de Linha
plt.figure(figsize=(20, 4))
plt.plot(iris_data['PetalaComp'], marker='o', lw=2, ls=':', color='black')
plt.grid(True)
plt.title('Dataset iris Comprimento da Petala')
plt.ylabel('PetalaCom (cm)')
```

```python
# Um Histograma
plt.figure(figsize=(4, 4))
plt.hist(x=iris_data['PetalaComp'], bins=8, linewidth=0.5, edgecolor="white")
plt.grid(True)
plt.title('Histograma PetalaLarg')
plt.xlabel('PetalaComp (cm)')
```

```python
fig, ax = plt.subplots(figsize=(4,4))

ax.scatter(x=iris_data['PetalaComp'], y=iris_data['PetalaLarg'], s=10, c='blue', marker='^')
ax.scatter(x=iris_data['SepalaComp'], y=iris_data['SepalaLarg'], s=50, c='green', marker='v', alpha=0.5)

plt.show()
```

```python
iris_data.Especie.unique()
```

```python
# Fazendo um scatter plot interessante
fig, ax = plt.subplots(figsize=(4,4))
x = 'PetalaComp'
y = 'PetalaLarg'
x1 = iris_data.loc[iris_data.Especie=='Iris-setosa',x]
y1 = iris_data.loc[iris_data.Especie=='Iris-setosa',y]
x2 = iris_data.loc[iris_data.Especie=='Iris-versicolor',x]
y2 = iris_data.loc[iris_data.Especie=='Iris-versicolor',y]
x3 = iris_data.loc[iris_data.Especie=='Iris-virginica',x]
y3 = iris_data.loc[iris_data.Especie=='Iris-virginica',y]
ax.scatter(x=x1, y=y1,c='blue', marker='^')
ax.scatter(x=x2, y=y2,c='orange', marker='o')
ax.scatter(x=x3, y=y3,c='green', marker='D')

plt.show()
```

```python
# Fazendo um segundo scatter plot interessante
fig, ax = plt.subplots(figsize=(4,4))
x = 'SepalaComp'
y = 'SepalaLarg'
x1 = iris_data.loc[iris_data.Especie=='Iris-setosa',x]
y1 = iris_data.loc[iris_data.Especie=='Iris-setosa',y]
x2 = iris_data.loc[iris_data.Especie=='Iris-versicolor',x]
y2 = iris_data.loc[iris_data.Especie=='Iris-versicolor',y]
x3 = iris_data.loc[iris_data.Especie=='Iris-virginica',x]
y3 = iris_data.loc[iris_data.Especie=='Iris-virginica',y]
ax.scatter(x=x1, y=y1,c='blue', marker='^', label='Setosa')
ax.scatter(x=x2, y=y2,c='orange', marker='o', label='Versicolor')
ax.scatter(x=x3, y=y3,c='green', marker='D', label='Virginica')
ax.legend()
plt.show()
```

# Então temos o Seaborn

O matplotlin as vezes pode ser meio complicado para programar, quando temos gráficos mais complexos! E então surgiu o [seaborn](https://seaborn.pydata.org/index.html)

## Vamos tentar o Seaborn!
Seaborn tem muitas funções interessantes para plotagem. Você pode criar jointplot, hisplot, pairplot, boxplot, violinplot (é muito legal), heatmap plot etc etc. Eu recomendo seaborn pessoalmente porque ele tem muitas técnicas legais de plotagem!

```python
iris_data.columns
```

```python
import seaborn as sns

sns.jointplot(x='SepalaComp', y = 'SepalaLarg', data = iris_data)
```

```python
#seaborn gráfico de dispersão por espécie em sépala_comprimento vs sépala_largura
sns.relplot(data=iris_data, x='SepalaComp', y = 'SepalaLarg', hue="Especie")
```

```python
sns.displot(iris_data, x='SepalaComp', hue="Especie")
```

```python
sns.displot(iris_data, x='SepalaLarg', hue="Especie")
```
