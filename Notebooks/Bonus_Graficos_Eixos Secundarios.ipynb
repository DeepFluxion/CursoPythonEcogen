```python
# importando as bibliotecas
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Criando dados de teste
t = np.arange(0.01, 10.0, 0.01) #dados de 0.01 at√© 10 com intervalo de 0.01
data1 = np.exp(t) # logaritmo dos valores de t
data2 = np.sin(2 * np.pi * t) # seno dos valores de 2*pi*t
```

```python
# Visualizando os dados data 1 separdamente

plt.plot(t,data1)
plt.xlabel('Valores time (s)')
plt.ylabel('Valores de data 1')
plt.show()
```

```python
# Visualizando os dados data 2 separdamente

plt.plot(t,data2)
plt.xlabel('Valores time (s)')
plt.ylabel('Valores de data 2')
plt.show()
```

```python
fig, ax1 = plt.subplots() #cria a figura

color = 'red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('data 1', color=color) # ajusta o titulo do eixos y
ax1.plot(t, data1, color=color) # plota a linha no axis 1
ax1.tick_params(axis='y', labelcolor=color) # ajuste dos parametros do axis1

ax2 = ax1.twinx()  # cria um segundo axis que compartilha o eixo x de axis1

color = 'tab:blue'
ax2.set_ylabel('data 2', color=color)  # nos j√° informamos os valores do titulo do eixo x no axis 1
ax2.plot(t, data2, color=color) # plota a linha no axis 2
ax2.tick_params(axis='y', labelcolor=color) # ajuste dos parametros do axis2

fig.tight_layout()  # melhora a apresneta√ß√£o do gr√°fico 
plt.show()
```
